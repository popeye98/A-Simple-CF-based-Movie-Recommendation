```python
import pandas as pd
from scipy import sparse
from sklearn.metrics.pairwise import cosine_similarity
```

```python
path='YOUR PATH'
```

```python
ratings=pd.read_csv(path+'/toy.csv',index_col=0)
```

```python
ratings.fillna(0, inplace=True)
ratings  #take this as out test data set
```

```python
ratings  # handling missing value 
```

```python

def standardize(row):
    new_row = (row - row.mean())/(row.max()-row.min())
    return new_row

df_std = ratings.apply(standardize).T    # as we are doing item to item CF so taking transpose 
                                            #to make row contains item not user
print(df_std)

sparse_df = sparse.csr_matrix(df_std.values)
corrMatrix = pd.DataFrame(cosine_similarity(sparse_df),index=ratings.columns,columns=ratings.columns)
corrMatrix
```

```python
corrMatrix = ratings.corr(method='pearson')
corrMatrix.head(6)
```

```python
def get_similar(movie_name,rating):
    similar_score = corrMatrix[movie_name]*(rating-2.5)  # -2.5 is due to handle rating 1 also 
                                                        # as rating 1 is there so user may hate this so not to predict
    similar_score = similar_score.sort_values(ascending=False)
    #print(type(similar_ratings))
    return similar_score
```

```python
action_lover = [("action 1",5),("romantic 2",1),("romantic3",1)]    # a user data
similar_scores = pd.DataFrame()
for movie,rating in action_lover:
    similar_scores = similar_scores.append(get_similar(movie,rating),ignore_index = True)

similar_scores.head(10)
```

```python
similar_scores.sum().sort_values(ascending=False)   # sum due to all his activity taking in consideration 
                                                          # then predicting final output  
```

```python
ratings=pd.merge(ratings,movies,on='movieId')
```

```python
ratings=ratings.drop(['timestamp','genres'],axis=1)  # droping unnecessary columns
```

```python
ratings
```

now making tale as 2d matix whose x axis is movies and y axis is user 

```python
userRatings = ratings.pivot_table(index=['userId'],columns=['title'],values='rating')
```

```python
userRatings.head()
```

creating threshold for better prediction and dimension reduction

```python
userRatings = userRatings.dropna(thresh=10, axis=1).fillna(0,axis=1)
```

```python
userRatings
```

```python
# normalising
corrMatrix=userRatings.corr('pearson')
corrMatrix
```

```python
romantic_lover = [("(500) Days of Summer (2009)",5),("Alice in Wonderland (2010)",3),("Aliens (1986)",1),("2001: A Space Odyssey (1968)",2)]
similar_movies = pd.DataFrame()
for movie,rating in romantic_lover:
    similar_movies = similar_movies.append(get_similar(movie,rating),ignore_index = True)

similar_movies.head(10)
```
